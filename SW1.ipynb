```python

from abc import ABC, abstractmethod
from datetime import datetime

# Abstract Base Class (ABC) for Employee
class Employee(ABC):
    def __init__(self, emp_id, emp_name, emp_address):
        self._emp_id = emp_id
        self._emp_name = emp_name
        self._emp_address = emp_address

    @property
    def emp_id(self):
        return self._emp_id

    @property
    def emp_name(self):
        return self._emp_name

    @property
    def emp_address(self):
        return self._emp_address

    @abstractmethod
    def calculate_salary(self):
        pass

# Fulltime Employee class inheriting from Employee
class FulltimeEmployee(Employee):
    def __init__(self, emp_id, emp_name, emp_address, allowance, rate):
        super().__init__(emp_id, emp_name, emp_address)
        self._allowance = allowance
        self._rate = rate

    @property
    def allowance(self):
        return self._allowance

    @property
    def rate(self):
        return self._rate

    def calculate_salary(self):
        return self._allowance + (self._rate * 8)  # assuming 8 hours of work per day

# PartTime Employee class inheriting from Employee
class PartTimeEmployee(Employee):
    def __init__(self, emp_id, emp_name, emp_address, rate):
        super().__init__(emp_id, emp_name, emp_address)
        self._rate = rate

    @property
    def rate(self):
        return self._rate

    def calculate_salary(self):
        return self._rate * 4  # assuming 4 hours of work per day

# Salary class
class Salary:
    def __init__(self, salary_id, salary, cut_off_date, days_of_work):
        self._salary_id = salary_id
        self._salary = salary
        self._cut_off_date = cut_off_date
        self._days_of_work = days_of_work

    @property
    def salary_id(self):
        return self._salary_id

    @property
    def salary(self):
        return self._salary

    @property
    def cut_off_date(self):
        return self._cut_off_date

    @property
    def days_of_work(self):
        return self._days_of_work

# Example usage
emp1 = FulltimeEmployee(1, "John Doe", "123 Main St", 500, 20)
emp2 = PartTimeEmployee(2, "Jane Doe", "456 Elm St", 15)

sal1 = Salary(1, emp1.calculate_salary(), datetime(2023, 3, 15), 20)
sal2 = Salary(2, emp2.calculate_salary(), datetime(2023, 3, 15), 10)

print("Employee 1 Salary:", sal1.salary)
print("Employee 2 Salary:", sal2.salary)
```
